````python
import matplotlib.pyplot as plt
if platform.system() == "Windows":
    plt.rcParams['font.family'] ='Malgun Gothic'
elif platform.system() == "Darwin":
    plt.rcParams['font.family'] ='AppleGothic'
elif platform.system() == "Linux":
    plt.rcParams["font.family"] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] =False
````

````python
jj_df.plot(x='관측일자(OBSR_YMD)', y='임계주파수F2층값(FO_F2_VALUE)', figsize=(12, 4));
````

````python
jj_df_quite_22 = jj_df[(jj_df['관측일자(OBSR_YMD)'] >= '2024-03-22') & (jj_df['관측일자(OBSR_YMD)'] < '2024-03-23')]
jj_df_storm = jj_df[(jj_df['관측일자(OBSR_YMD)'] >= '2024-03-24') & (jj_df['관측일자(OBSR_YMD)'] < '2024-03-26')]
````

````python
jj_df_quite_22 = jj_df_quite_22.sort_values(by='관측일자(OBSR_YMD)')
jj_df_quite_22 = jj_df_quite_22.reset_index(drop=True)
jj_df_quite_22
````

````python
jj_df_quite_copy = jj_df_quite_22.copy()
jj_df_quite_copy['관측일자(OBSR_YMD)'] += pd.DateOffset(days=1)
jj_df_quite_copy
````

````python
jj_df_quite = pd.concat([jj_df_quite_22, jj_df_quite_copy]).reset_index(drop=True)
jj_df_quite
````

````python
jj_df_quite['offset'] = jj_df_quite['관측일자(OBSR_YMD)'] - jj_df_quite['관측일자(OBSR_YMD)'].min()
jj_df_quite
````

````python
jj_df_quite['offset'].dt.total_seconds()
````

````python
import datetime
import matplotlib.ticker as ticker
````

````python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.plot(jj_df_quite['offset'].dt.total_seconds(), jj_df_quite['임계주파수F2층값(FO_F2_VALUE)'], 'k-', label='quite')
ax.set_ylabel('임계주파수F2층값(FO_F2_VALUE)')
locator = ticker.MultipleLocator(base=12*3600)
ax.xaxis.set_major_locator(locator)
formatter = ticker.FuncFormatter(lambda x, _: str(datetime.timedelta(seconds=x)))                                                                                                                                                                                                                         
ax.xaxis.set_major_formatter(formatter)
ax.legend()
plt.show()
````

````python
jj_df_storm = jj_df_storm.sort_values(by='관측일자(OBSR_YMD)')
jj_df_storm = jj_df_storm.reset_index(drop=True)
jj_df_storm['offset'] = jj_df_storm['관측일자(OBSR_YMD)'] - jj_df_storm['관측일자(OBSR_YMD)'].min()
jj_df_storm
````

````python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.plot(jj_df_quite['offset'].dt.total_seconds(), jj_df_quite['임계주파수F2층값(FO_F2_VALUE)'], 'k-', label='quite')
ax.plot(jj_df_storm['offset'].dt.total_seconds(), jj_df_storm['임계주파수F2층값(FO_F2_VALUE)'], 'r-', label='storm')

ax.xaxis.set_major_locator(ticker.MultipleLocator(base=12*3600))
formatter = ticker.FuncFormatter(lambda x, _: str(datetime.timedelta(seconds=x))[:-3])                                                                                                                                                                                                                         
ax.xaxis.set_major_formatter(formatter)
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=2.4*3600))

ax.yaxis.set_major_locator(ticker.MultipleLocator(base=2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(base=0.5))

# ax.set_xlim(jj_df_quite['offset'].dt.total_seconds().min(), jj_df_quite['offset'].dt.total_seconds().max())
ax.set_ylim(2, 16)

ax.tick_params(axis='both', which='minor', length=5, direction='in', labelsize=15)
ax.tick_params(axis='both', which='major', length=12, direction='in', labelsize=15)
ax.set_ylabel('foF2 (MHz)', fontsize=15)
ax.set_title('Jeju Ionosonde foF2 plot (final)', fontsize=20)
ax.legend(prop={'size': 15})
plt.show()
````

### Icheon

````python
ic = pd.read_csv("data/Ionosonde/Korea_20240322_20240325_icheon.csv")
ic["관측일자(OBSR_YMD)"] = pd.to_datetime(ic["관측일자(OBSR_YMD)"], format="%Y-%m-%d %H:%M:%S")
ic_df = ic[['관측일자(OBSR_YMD)', '임계주파수F2층값(FO_F2_VALUE)', 'F2층높이(F2_HG)']]
ic_df.describe()
````

````python
ic_df.plot(x='관측일자(OBSR_YMD)', y='임계주파수F2층값(FO_F2_VALUE)', figsize=(12, 4));
````

````python
ic_df_quite_22 = ic_df[(ic_df['관측일자(OBSR_YMD)'] >= '2024-03-22') & (ic_df['관측일자(OBSR_YMD)'] < '2024-03-23')]
ic_df_quite_22 = ic_df_quite_22.sort_values(by='관측일자(OBSR_YMD)')
ic_df_quite_22 = ic_df_quite_22.reset_index(drop=True)
ic_df_quite_copy = ic_df_quite_22.copy()
ic_df_quite_copy['관측일자(OBSR_YMD)'] += pd.DateOffset(days=1)
ic_df_quite = pd.concat([ic_df_quite_22, ic_df_quite_copy]).reset_index(drop=True)
ic_df_quite['offset'] = ic_df_quite['관측일자(OBSR_YMD)'] - ic_df_quite['관측일자(OBSR_YMD)'].min()

ic_df_storm = ic_df[(ic_df['관측일자(OBSR_YMD)'] >= '2024-03-24') & (ic_df['관측일자(OBSR_YMD)'] < '2024-03-26')]
ic_df_storm = ic_df_storm.sort_values(by='관측일자(OBSR_YMD)')
ic_df_storm = ic_df_storm.reset_index(drop=True)
ic_df_storm['offset'] = ic_df_storm['관측일자(OBSR_YMD)'] - ic_df_storm['관측일자(OBSR_YMD)'].min()


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.plot(ic_df_quite['offset'].dt.total_seconds(), ic_df_quite['임계주파수F2층값(FO_F2_VALUE)'], 'k-', label='quite')
ax.plot(ic_df_storm['offset'].dt.total_seconds(), ic_df_storm['임계주파수F2층값(FO_F2_VALUE)'], 'r-', label='storm')

ax.xaxis.set_major_locator(ticker.MultipleLocator(base=12*3600))
formatter = ticker.FuncFormatter(lambda x, _: str(datetime.timedelta(seconds=x))[:-3])                                                                                                                                                                                                                         
ax.xaxis.set_major_formatter(formatter)
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=2.4*3600))

ax.yaxis.set_major_locator(ticker.MultipleLocator(base=2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(base=0.5))

ax.set_ylim(2, 16)

ax.tick_params(axis='both', which='minor', length=5, direction='in', labelsize=15)
ax.tick_params(axis='both', which='major', length=12, direction='in', labelsize=15)
ax.set_ylabel('foF2 (MHz)', fontsize=15)
ax.set_title('Icheon Ionosonde foF2 plot (final)', fontsize=20)
ax.legend(prop={'size': 15})
plt.show()
````
