# Kp Index

https://kp.gfz-potsdam.de/en/

[Python Client](https://kp.gfz-potsdam.de/app/webservice/python)

```python
from getKpindex import getKpindex

# 2011 - 2024 Kp index data
starttime = "2011-01-01"
endtime = "2024-12-31"

time, index, status = getKpindex(starttime, endtime, "Kp", "def")
```

```python
len(time), len(index), len(status)
```

```python
import pandas as pd

df = pd.DataFrame({"time": time, "Kp index": index, "status": status})
```

```python
df
```

```python
# G4

g4_event = (df["Kp index"] > 7) & (df["Kp index"] < 9)
sum(g4_event.values)
```

```python
df[g4_event]
```

```python
df['obstime'] = pd.to_datetime(df["time"], utc=True)
df['date'] = df['obstime'].dt.date
df
```

```python
grouped = df[g4_event].groupby("date")
dates = grouped.size()
dates
```

```python
len(dates)
```

```python
df['obstime'][0]
```

```python
import matplotlib.pyplot as plt
```

```python
incides = dates.index
incides
```

```python
for idx in incides:
    cond1 = df['date'] > idx - pd.Timedelta(days=3.5)
    cond2 = df['date'] < idx + pd.Timedelta(days=3.5)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    df[cond1 & cond2].plot(x="obstime", y="Kp index", kind="bar", ax=ax)
    plt.show()
```
