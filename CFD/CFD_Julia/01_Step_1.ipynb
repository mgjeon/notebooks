```julia
using Plots
```

```julia
x_end = 2
nx = 81  
Δx = x_end / (nx-1)
x = range(0, stop=x_end, length=nx)

t_end = 0.625
nt = 25    
Δt = t_end / nt
t = range(0, stop=t_end, length=nt)

c = 1      #assume wavespeed of c = 1

print(Δt)
```

```julia
u⁰ = ones(nx)
u⁰[0.5 .≤ x .≤ 1] .= 2  # Note use of . (dot) broadcasting syntax
print(u⁰)
```

```julia
plot(x, u⁰)
```

```julia
uⁿ⁺¹ = copy(u⁰)

for n in 1:nt
    uⁿ = copy(uⁿ⁺¹)
    for i in 2:nx
        uⁿ⁺¹[i] = uⁿ[i] - c * Δt/Δx * (uⁿ[i] - uⁿ[i-1])
    end
end

u = uⁿ⁺¹;
```

```julia
plot(x, u)
```
